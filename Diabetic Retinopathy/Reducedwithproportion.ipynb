```python
import os
import shutil
import random
```

```python
# Defining the paths to the original dataset and the destination folder for the reduced_dataset
original_dataset_path = 'data/train'
reduced_dataset_path = 'reduceddata/train'
class_folders = os.listdir(original_dataset_path)
```

```python
# Determine the class with the fewest images to establish the target number
min_images = float('inf')

for class_folder in class_folders:
    class_path = os.path.join(original_dataset_path, class_folder)
    image_files = os.listdir(class_path)
    num_images = len(image_files)
    
    if num_images < min_images:
        min_images = num_images

# Percentage of data to keep (e.g., 35%)
percentage_to_keep = 60
```

```python
# Iterating through each class folder
for class_folder in class_folders:
    class_path = os.path.join(original_dataset_path, class_folder)
    reduced_class_path = os.path.join(reduced_dataset_path, class_folder)
    
    # Create the corresponding class folder in the reduced dataset
    os.makedirs(reduced_class_path, exist_ok=True)
    
    # Listing all image files in the class folder
    image_files = os.listdir(class_path)
    
    # Calculate the number of images to keep based on the class with the fewest images
    num_to_keep = int(min_images * (percentage_to_keep / 100.00))
    
    # Randomly shuffle the list of image files
    random.shuffle(image_files)
    
    # Copy the selected number of images to the reduced dataset folder
    for i in range(num_to_keep):
        src_image_path = os.path.join(class_path, image_files[i])
        dest_image_path = os.path.join(reduced_class_path, image_files[i])
        shutil.copy(src_image_path, dest_image_path)

# The reduced dataset is created with balanced class proportions
```

```python
import os
import random
import shutil

# Paths to the original train and test directories
train_dir = 'train'
test_dir = 'test'

# Create new directories for the smaller balanced train and test sets
small_balanced_train_dir = 'small_balanced_train'
small_balanced_test_dir = 'small_balanced_test'

# Create the smaller balanced train and test directories if they don't exist
os.makedirs(small_balanced_train_dir, exist_ok=True)
os.makedirs(small_balanced_test_dir, exist_ok=True)

# Function to copy a specified number of random images from the source to destination
def copy_random_images(source_dir, destination_dir, num_images):
    image_files = os.listdir(source_dir)
    random.shuffle(image_files)  # Shuffle the images randomly
    
    for i, image in enumerate(image_files):
        if i >= num_images:
            break  # Stop copying once the desired number is reached
        
        src_path = os.path.join(source_dir, image)
        dest_path = os.path.join(destination_dir, image)
        shutil.copy(src_path, dest_path)

# Set the desired number of images per class
num_images_per_class = 5000  # Adjust this number as needed

# Copy a specified number of random images from both 'DR' and 'No_DR' subfolders
copy_random_images(os.path.join(train_dir, 'DR'), os.path.join(small_balanced_train_dir, 'DR'), num_images_per_class)
copy_random_images(os.path.join(train_dir, 'No_DR'), os.path.join(small_balanced_train_dir, 'No_DR'), num_images_per_class)
copy_random_images(os.path.join(test_dir, 'DR'), os.path.join(small_balanced_test_dir, 'DR'), num_images_per_class)
copy_random_images(os.path.join(test_dir, 'No_DR'), os.path.join(small_balanced_test_dir, 'No_DR'), num_images_per_class)

print(f"{num_images_per_class} images from each class copied to smaller balanced datasets.")

```
